# Test Modeller for SPY Option Calls

```python
# Load in data
import pandas as pd

DATA_PATH='data/raw.csv'

data = pd.read_csv(DATA_PATH)
data = data.dropna()
data
```

```python
import numpy as np
# Create a new column that contains the combined values from the last 4 columns
DATA_LABEL = 'multi_target_label'
data[DATA_LABEL] = data.iloc[:, -4:].apply(lambda row: np.array(row), axis=1)
data
```

```python
import base64
import dateutil.parser

BASE64_COLUMN_NAME='text'
TEXT_COLUMN_NAME='text'
data[TEXT_COLUMN_NAME] = data[BASE64_COLUMN_NAME].apply(lambda x: base64.b64decode(x).decode())
data
```

```python
TRAIN_TEST_RATIO = 0.2

def split_dataset(X, y, test_ratio):
    shuffled_indices = np.random.permutation(len(X))
    test_set_size = int(len(X) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return X.iloc[train_indices], y.iloc[train_indices].to_numpy(), X.iloc[test_indices], y.iloc[test_indices].to_numpy()

X_train, t_train, X_test, t_test = split_dataset(data, data[DATA_LABEL], TRAIN_TEST_RATIO)

print(f"X_train.shape:\t{X_train.shape}")
print(f"t_train.shape:\t{t_train.shape}")
print(f"X_test.shape:\t{X_test.shape}")
print(f"t_test.shape:\t{t_test.shape}")
```

```python
t_train.shape
t_train_reshaped = np.vstack(t_train)
t_train_reshaped.shape
```

```python
t_test_reshaped = np.vstack(t_test)
t_test_reshaped.shape
```

```python
from sklearn.pipeline import Pipeline, FunctionTransformer
from sklearn.compose import ColumnTransformer
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.multioutput import MultiOutputRegressor, RegressorChain
from sklearn.svm import LinearSVR, SVR

# Feature Extraction
text_preprocess = Pipeline([
    ('count', CountVectorizer()),
    ('transform', TfidfTransformer())
])

# Combine text feature extraction with day_of_week feature
ct = ColumnTransformer([
    ('text', text_preprocess, 'text'),
    ('dow', FunctionTransformer(lambda x: date_to_day_of_week(x)), ['date'])
])

def date_to_day_of_week(x):
  x['date'] = x['date'].apply(lambda row: dateutil.parser.parse(row).weekday())
  return x

# Add the final regressor to the pipeline
pipe = Pipeline([
    ('features', ct),
    ('reg', RegressorChain(base_estimator=SVR()))
])

#pipe.fit(X_train, t_train_reshaped)
```

```python
from sklearn.metrics import make_scorer

def mse(y_true, y_pred):
  diff = np.square(y_true - y_pred)
  return np.mean(diff, axis=0)

def mse_scorer(y_true, y_pred):
  diff = np.square(y_true - y_pred)
  return np.mean(diff)

score = make_scorer(mse_scorer, greater_is_better=False)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor, HistGradientBoostingRegressor

ANALYSERS = ['char_wb']
NGRAMS = [(1,14)]
#ESTIMATORS = [LinearSVR(dual='auto', C=1), SVR(C=1), SVR(C=10), SVR(C=100), SVR(C=1000)]
ESTIMATORS = [RandomForestRegressor()]

param_grid = [
    {
        "features__text__count__ngram_range": NGRAMS,
        "features__text__count__analyzer": ANALYSERS,
        "reg__base_estimator": ESTIMATORS
    }
]
grid = GridSearchCV(pipe, param_grid=param_grid, scoring=score)
grid.fit(X_train, t_train_reshaped)
grid.best_estimator_.get_params()
```

```python
model = grid.best_estimator_
"""
SVR Early Results
Test Mean Squared Error:	[15.16706203 15.0216847  39.14198806 51.98481767]
Test Score:	-30.328888114477042
Train Mean Squared Error:	[0.0093681  0.00968157 0.00983533 0.01000529]
Train Score:	-0.009722570035954066
"""
print(f"Test Mean Squared Error:\t{mse(t_test_reshaped, model.predict(X_test))}")
print(f"Test Score:\t{score(model, X_test, t_test_reshaped)}")
print(f"Train Mean Squared Error:\t{mse(t_train_reshaped, model.predict(X_train))}")
print(f"Train Score:\t{score(model, X_train, t_train_reshaped)}")
```

```python
cv = model['count'].transform(predict_data1)
tf_idf_vector = model['transform'].transform(cv)
feature_names = model['count'].get_feature_names_out()
feature_names
```

```python
def sort_coo(coo_matrix):
  tuples = zip(coo_matrix.col, coo_matrix.data)
  return sorted(tuples, key= lambda x: (x[1],x[0]), reverse=True)
sorted_items = sort_coo(tf_idf_vector.tocoo())
```

```python
def extract_topn_from_vector(feature_names, sorted_items, topn=7):
  sorted_items = sorted_items[:topn]
  score_vals = []
  feature_vals = []
  for idx, score in sorted_items:
    score_vals.append(round(score,3))
    feature_vals.append(feature_names[idx])
  results = {}
  for idx in range(len(feature_vals)):
    results[feature_vals[idx]] = score_vals[idx]
  return results
extract_topn_from_vector(feature_names, sorted_items)
```
